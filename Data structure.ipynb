```python
#Assignments: Data Structures in Python
#Q1.write a code to reverse a string.
#1. Using Slicing
# Original string
original_string = "hello"

# Reverse the string using slicing
reversed_string = original_string[::-1]

print(reversed_string)
2. Using a Loop
# Original string
original_string = "hello"

# Reverse the string using a loop
reversed_string = ""
for char in original_string:
    reversed_string = char + reversed_string

print(reversed_string)
3. Using reversed() Function
# Original string
original_string = "hello"

# Reverse the string using reversed() and join()
reversed_string = ''.join(reversed(original_string))

print(reversed_string)

#Q2.write a code to count the no. of vowel in astring.
1. Using a Loop
# Original string
string = "hello world"

# Vowels to check
vowels = "aeiou"

# Initialize the count
vowel_count = 0

# Count the vowels using a loop
for char in string:
    if char.lower() in vowels:
        vowel_count += 1

print(vowel_count)
2. Using List Comprehension
# Original string
string = "hello world"

# Vowels to check
vowels = "aeiou"

# Count the vowels using list comprehension
vowel_count = sum(1 for char in string if char.lower() in vowels)

print(vowel_count)
3. Using collections.Counter
from collections import Counter

# Original string
string = "hello world"

# Vowels to check
vowels = "aeiou"

# Count characters using Counter
char_count = Counter(string.lower())

# Sum counts of vowels
vowel_count = sum(char_count[vowel] for vowel in vowels)

print(vowel_count)
4. Using Regular Expressions
import re

# Original string
string = "hello world"

# Find all vowels using regular expressions
vowel_count = len(re.findall(r'[aeiou]', string, re.IGNORECASE))

print(vowel_count)
#Q3.write a code to check if a given string is a palindrome or not.
1. Using Slicing
# Function to check if a string is a palindrome
def is_palindrome(s):
    # Remove spaces and convert to lowercase for a case-insensitive comparison
    s = s.replace(" ", "").lower()
    return s == s[::-1]

# Test cases
print(is_palindrome("A man a plan a canal Panama"))
print(is_palindrome("hello"))
2. Using a Loop
# Function to check if a string is a palindrome using a loop
def is_palindrome(s):
    # Remove spaces and convert to lowercase
    s = s.replace(" ", "").lower()
    left, right = 0, len(s) - 1

    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1

    return True

# Test cases
print(is_palindrome("A man a plan a canal Panama"))
print(is_palindrome("hello"))
3. Using Recursion
# Function to check if a string is a palindrome using recursion
def is_palindrome(s):
    # Remove spaces and convert to lowercase
    s = s.replace(" ", "").lower()

    def check_palindrome(left, right):
        if left >= right:
            return True
        if s[left] != s[right]:
            return False
        return check_palindrome(left + 1, right - 1)

    return check_palindrome(0, len(s) - 1)

# Test cases
print(is_palindrome("A man a plan a canal Panama"))
print(is_palindrome("hello"))
4. Using Regular Expressions
import re

# Function to check if a string is a palindrome using regular expressions
def is_palindrome(s):
    # Remove non-alphanumeric characters and convert to lowercase
    s = re.sub(r'[^a-zA-Z0-9]', '', s).lower()
    return s == s[::-1]

# Test cases
print(is_palindrome("A man a plan a canal Panama"))
print(is_palindrome("hello"))

#Q4.write a code to check if 2 string are anagrams of each others.
1. Using Sorting
# Function to check if two strings are anagrams using sorting
def are_anagrams(str1, str2):
    # Remove spaces and convert to lowercase for a case-insensitive comparison
    str1 = str1.replace(" ", "").lower()
    str2 = str2.replace(" ", "").lower()

    return sorted(str1) == sorted(str2)

# Test cases
print(are_anagrams("listen", "silent"))
print(are_anagrams("hello", "world"))
2. Using collections.Counter
from collections import Counter

# Function to check if two strings are anagrams using Counter
def are_anagrams(str1, str2):
    # Remove spaces and convert to lowercase for a case-insensitive comparison
    str1 = str1.replace(" ", "").lower()
    str2 = str2.replace(" ", "").lower()

    return Counter(str1) == Counter(str2)

# Test cases
print(are_anagrams("listen", "silent"))
print(are_anagrams("hello", "world"))
3. Using a Frequency Dictionary
# Function to check if two strings are anagrams using a frequency dictionary
def are_anagrams(str1, str2):
    # Remove spaces and convert to lowercase
    str1 = str1.replace(" ", "").lower()
    str2 = str2.replace(" ", "").lower()

    # Create frequency dictionaries
    freq1 = {}
    freq2 = {}

    for char in str1:
        freq1[char] = freq1.get(char, 0) + 1

    for char in str2:
        freq2[char] = freq2.get(char, 0) + 1

    return freq1 == freq2

# Test cases
print(are_anagrams("listen", "silent"))
print(are_anagrams("hello", "world"))
4. Using Regular Expressions
import re
from collections import Counter

# Function to check if two strings are anagrams using regular expressions
def are_anagrams(str1, str2):
    # Remove non-alphanumeric characters and convert to lowercase
    str1 = re.sub(r'[^a-zA-Z0-9]', '', str1).lower()
    str2 = re.sub(r'[^a-zA-Z0-9]', '', str2).lower()

    return Counter(str1) == Counter(str2)

# Test cases
print(are_anagrams("Listen!!!", "Silent"))
print(are_anagrams("hello", "world"))

#Q5.write a code tofind all occurrence of a given substring within another string.
1. Using str.find() in a Loop
def find_all_occurrences(s, substring):
    start = 0
    occurrences = []

    while True:
        start = s.find(substring, start)
        if start == -1:
            break
        occurrences.append(start)
        start += len(substring)  # Move past the last found substring

    return occurrences

# Test cases
string = "This is a test. This test is only a test."
substring = "test"

print(find_all_occurrences(string, substring))
2. Using Regular Expressions
import re

def find_all_occurrences(s, substring):
    pattern = re.escape(substring)  # Escape special characters in the substring
    matches = re.finditer(pattern, s)
    occurrences = [match.start() for match in matches]

    return occurrences

# Test cases
string = "This is a test. This test is only a test."
substring = "test"

print(find_all_occurrences(string, substring))
3. Using List Comprehension
def find_all_occurrences(s, substring):
    return [i for i in range(len(s)) if s.startswith(substring, i)]

# Test cases
string = "This is a test. This test is only a test."
substring = "test"

print(find_all_occurrences(string, substring))
4. Using str.index() in a Loop
def find_all_occurrences(s, substring):
    start = 0
    occurrences = []

    while True:
        try:
            start = s.index(substring, start)
            occurrences.append(start)
            start += len(substring)
        except ValueError:
            break

    return occurrences

# Test cases
string = "This is a test. This test is only a test."
substring = "test"

print(find_all_occurrences(string, substring))

#Q6.write a code to perform basic string compression using the counts of repeated characters.
def compress_string(s):
    # Initialize the compressed string and a count variable
    compressed = []
    count = 1

    # Iterate over the string
    for i in range(1, len(s)):
        # If the current character is the same as the previous one, increment the count
        if s[i] == s[i - 1]:
            count += 1
        else:
            # Otherwise, append the previous character and count to the compressed string
            compressed.append(s[i - 1] + str(count))
            count = 1  # Reset count for the new character

    # Append the last character and its count
    compressed.append(s[-1] + str(count))

    # Join the list into a string
    compressed_str = ''.join(compressed)

    # Return the original string if the compressed string is not shorter
    return compressed_str if len(compressed_str) < len(s) else s

# Example usage:
input_string = "aabcccccaaa"
compressed_string = compress_string(input_string)
print(compressed_string)

#Q7.write acode to determine if a string has all unique character.
def has_all_unique_characters(s):
    # Initialize an empty set to store characters
    seen = set()

    # Iterate over each character in the string
    for char in s:
        # If the character is already in the set, it's a duplicate
        if char in seen:
            return False
        # Add the character to the set
        seen.add(char)

    # If no duplicates were found, return True
    return True

# Example usage:
input_string = "abcdef"
print(has_all_unique_characters(input_string))

input_string = "aabbcc"
print(has_all_unique_characters(input_string))

#Q8.write a code to convert a given string  to uppercase or lowercase.
def convert_case(s, to_upper=True):
    """
    Convert the given string to uppercase or lowercase.

    :param s: The string to be converted.
    :param to_upper: If True, convert to uppercase; if False, convert to lowercase.
    :return: The converted string.
    """
    if to_upper:
        return s.upper()
    else:
        return s.lower()

# Example usage:
input_string = "Hello World!"

# Convert to uppercase
uppercase_string = convert_case(input_string, to_upper=True)
print(uppercase_string)

# Convert to lowercase
lowercase_string = convert_case(input_string, to_upper=False)
print(lowercase_string)

#Q9.write a code to count a no. of words in a string.
def count_words(s):
    """
    Count the number of words in the given string.

    :param s: The string in which words are to be counted.
    :return: The number of words in the string.
    """
    # Split the string into words based on whitespace
    words = s.split()

    # Return the number of words
    return len(words)

# Example usage:
input_string = "Hello world! This is a test string."
word_count = count_words(input_string)
print(word_count)

#Q10.write a code to concatenate 2 string without using the + operator.
def concatenate_strings(s1, s2):
    """
    Concatenate two strings without using the '+' operator.

    :param s1: The first string.
    :param s2: The second string.
    :return: The concatenated string.
    """
    # Use join() method to concatenate strings
    return ''.join([s1, s2])

# Example usage:
string1 = "Hello, "
string2 = "world!"
concatenated_string = concatenate_strings(string1, string2)
print(concatenated_string)

#Q11.write a code to remove all occurrence of a specific element from a list.
def remove_element(lst, element):
    """
    Remove all occurrences of a specific element from a list.

    :param lst: The list from which elements will be removed.
    :param element: The element to remove from the list.
    :return: A new list with all occurrences of the specified element removed.
    """
    # Use a list comprehension to create a new list without the specified element
    return [x for x in lst if x != element]

# Example usage:
my_list = [1, 2, 3, 2, 4, 2, 5]
element_to_remove = 2
new_list = remove_element(my_list, element_to_remove)
print(new_list)

#Q12.implement a code to find the second largest no. in a given list of integer.
def find_second_largest(lst):
    """
    Find the second largest number in a list of integers.

    :param lst: The list of integers.
    :return: The second largest number, or None if there is no such number.
    """
    # Check if the list has fewer than 2 unique elements
    if len(lst) < 2:
        return None

    # Initialize variables to store the largest and second largest numbers
    largest = second_largest = float('-inf')

    # Iterate over each number in the list
    for number in lst:
        # Update largest and second largest numbers
        if number > largest:
            second_largest = largest
            largest = number
        elif largest > number > second_largest:
            second_largest = number

    # Return the second largest number if it was updated, else None
    return second_largest if second_largest != float('-inf') else None

# Example usage:
numbers = [10, 20, 4, 45, 99, 99]
second_largest_number = find_second_largest(numbers)
print(second_largest_number)

#Q13.create a code to count the occurence of each element ina list & return a dictionary with elements as keys & their counts as values.
def count_occurrences(lst):
    """
    Count the occurrences of each element in a list and return a dictionary.

    :param lst: The list of elements to count.
    :return: A dictionary with elements as keys and their counts as values.
    """
    # Initialize an empty dictionary to store the counts
    counts = {}

    # Iterate over each element in the list
    for element in lst:
        # Update the count of the element in the dictionary
        if element in counts:
            counts[element] += 1
        else:
            counts[element] = 1

    return counts

# Example usage:
my_list = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
occurrence_counts = count_occurrences(my_list)
print(occurrence_counts)

#Q14.write a code to reverse a list in_place without using any built_in reserve functions.
def reverse_list(lst):
    """
    Reverse a list in place without using any built-in reverse functions.

    :param lst: The list to be reversed.
    """
    left = 0
    right = len(lst) - 1

    # Swap elements from both ends until the pointers meet
    while left < right:
        # Swap the elements
        lst[left], lst[right] = lst[right], lst[left]
        # Move the pointers towards the center
        left += 1
        right -= 1

# Example usage:
my_list = [1, 2, 3, 4, 5]
reverse_list(my_list)
print(my_list)

#Q15.implement a code to find & remove duplicates from a list while preserving the original order of elements.
def remove_duplicates(lst):
    """
    Remove duplicates from a list while preserving the original order of elements.

    :param lst: The list from which duplicates will be removed.
    :return: A new list with duplicates removed.
    """
    seen = set()       # Set to track seen elements
    result = []        # List to store the result with no duplicates

    for element in lst:
        if element not in seen:
            seen.add(element)  # Add the element to the set of seen elements
            result.append(element)  # Add the element to the result list

    return result

# Example usage:
my_list = [1, 2, 2, 3, 4, 4, 5, 1]
unique_list = remove_duplicates(my_list)
print(unique_list)

#Q16.create a code to check if given list is sorted (either in ascending or descending order) or not.
def is_sorted(lst):
    """
    Check if the given list is sorted in either ascending or descending order.

    :param lst: The list to check.
    :return: True if the list is sorted in ascending or descending order, False otherwise.
    """
    if not lst:  # An empty list is considered sorted
        return True

    ascending = descending = True

    # Check if list is sorted in ascending order
    for i in range(len(lst) - 1):
        if lst[i] > lst[i + 1]:
            ascending = False
            break

    # Check if list is sorted in descending order
    for i in range(len(lst) - 1):
        if lst[i] < lst[i + 1]:
            descending = False
            break

    return ascending or descending

# Example usage:
list1 = [1, 2, 3, 4, 5]
list2 = [5, 4, 3, 2, 1]
list3 = [1, 3, 2, 4, 5]
list4 = []

print(is_sorted(list1))
print(is_sorted(list2))
print(is_sorted(list3))
print(is_sorted(list4))

#Q17.write a code to merge 2 sorted list into asingle sorted list.
def merge_sorted_lists(list1, list2):
    """
    Merge two sorted lists into a single sorted list.

    :param list1: The first sorted list.
    :param list2: The second sorted list.
    :return: A merged list that is sorted.
    """
    merged_list = []
    i, j = 0, 0  # Pointers for list1 and list2

    # Iterate through both lists and merge them
    while i < len(list1) and j < len(list2):
        if list1[i] <= list2[j]:
            merged_list.append(list1[i])
            i += 1
        else:
            merged_list.append(list2[j])
            j += 1

    # Append remaining elements of list1 (if any)
    while i < len(list1):
        merged_list.append(list1[i])
        i += 1

    # Append remaining elements of list2 (if any)
    while j < len(list2):
        merged_list.append(list2[j])
        j += 1

    return merged_list

# Example usage:
sorted_list1 = [1, 3, 5, 7]
sorted_list2 = [2, 4, 6, 8]
merged_list = merge_sorted_lists(sorted_list1, sorted_list2)
print(merged_list)

#Q18.implement a code to find the intersection of 2 given list.
def intersection_of_lists(list1, list2):
    """
    Find the intersection of two lists.

    :param list1: The first list.
    :param list2: The second list.
    :return: A list containing elements that are present in both lists.
    """
    # Convert lists to sets to find common elements
    set1 = set(list1)
    set2 = set(list2)

    # Find the intersection of both sets
    intersection = set1 & set2

    # Convert the result back to a list and return
    return list(intersection)

# Example usage:
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]
common_elements = intersection_of_lists(list1, list2)
print(common_elements)

#Q19.create a code to find a union of 2 list without duplicates.
def union_of_lists(list1, list2):
    """
    Find the union of two lists without duplicates.

    :param list1: The first list.
    :param list2: The second list.
    :return: A list containing all unique elements from both lists.
    """
    # Convert lists to sets to remove duplicates and find the union
    set1 = set(list1)
    set2 = set(list2)

    # Find the union of both sets
    union = set1 | set2

    # Convert the result back to a list and return
    return list(union)

# Example usage:
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]
union_list = union_of_lists(list1, list2)
print(union_list)

#Q20.write a code to shuffle a given list randomly without using any built_in shuffle function.
import random

def shuffle_list(lst):
    """
    Shuffle a given list randomly without using built-in shuffle functions.

    :param lst: The list to be shuffled.
    """
    # Iterate over the list in reverse order
    for i in range(len(lst) - 1, 0, -1):
        # Generate a random index from 0 to i
        j = random.randint(0, i)
        # Swap elements at indices i and j
        lst[i], lst[j] = lst[j], lst[i]

# Example usage:
my_list = [1, 2, 3, 4, 5]
shuffle_list(my_list)
print(my_list)

#Q21.write a code to that takes 2 tuples as input & return a new tuple containing elements that are common to both input tuples.
def common_elements_tuple(tuple1, tuple2):
    """
    Find common elements between two tuples and return a new tuple with those elements.

    :param tuple1: The first tuple.
    :param tuple2: The second tuple.
    :return: A new tuple containing elements common to both input tuples.
    """
    # Convert tuples to sets to find common elements
    set1 = set(tuple1)
    set2 = set(tuple2)

    # Find the intersection of both sets
    common_elements = set1 & set2

    # Convert the set of common elements back to a tuple and return
    return tuple(common_elements)

# Example usage:
tuple1 = (1, 2, 3, 4, 5)
tuple2 = (4, 5, 6, 7, 8)
common_tuple = common_elements_tuple(tuple1, tuple2)
print(common_tuple)

#Q22.create a code that prompt the user to enter 2 sets of integer seprated by commas.then, print the intersection of these 2 sets.
def get_set_from_input(prompt):
    """
    Prompt the user for a set of integers separated by commas and return the set.

    :param prompt: The prompt message for user input.
    :return: A set of integers.
    """
    user_input = input(prompt)
    # Split the input string by commas and convert each part to an integer
    elements = map(int, user_input.split(','))
    # Return the set of elements
    return set(elements)

def main():
    """
    Main function to handle user input and print the intersection of two sets.
    """
    print("Enter two sets of integers separated by commas.")

    # Get the first set from the user
    set1 = get_set_from_input("Enter the first set of integers: ")

    # Get the second set from the user
    set2 = get_set_from_input("Enter the second set of integers: ")

    # Compute the intersection of the two sets
    intersection = set1 & set2

    # Print the intersection
    print("The intersection of the two sets is:", intersection)

# Run the main function
if __name__ == "__main__":
    main()

    #Q23.write a code to concatenate 2 tuples. the function should take 2 as input & return a new tuple containing element from both input tuples.
def concatenate_tuples(tuple1, tuple2):
    """
    Concatenate two tuples and return a new tuple containing elements from both tuples.

    :param tuple1: The first tuple.
    :param tuple2: The second tuple.
    :return: A new tuple containing elements from both input tuples.
    """
    # Concatenate the two tuples
    concatenated_tuple = tuple1 + tuple2

    return concatenated_tuple

# Example usage:
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
result = concatenate_tuples(tuple1, tuple2)
print(result)

#Q24.Develop a code that prompts  the user to input 2 sets of string.then,print the element that are present in the 1st set but not in the 2nd set.
def get_set_from_input(prompt):
    """
    Prompt the user for a set of strings separated by commas and return the set.

    :param prompt: The prompt message for user input.
    :return: A set of strings.
    """
    user_input = input(prompt)
    # Split the input string by commas and strip whitespace from each element
    elements = map(str.strip, user_input.split(','))
    # Return the set of elements
    return set(elements)

def main():
    """
    Main function to handle user input and print elements in the first set but not in the second set.
    """
    print("Enter two sets of strings separated by commas.")

    # Get the first set from the user
    set1 = get_set_from_input("Enter the first set of strings: ")

    # Get the second set from the user
    set2 = get_set_from_input("Enter the second set of strings: ")

    # Find elements present in the first set but not in the second set
    difference = set1 - set2

    # Print the result
    print("Elements in the first set but not in the second set:", difference)

# Run the main function
if __name__ == "__main__":
    main()


#Q25.create a code that takes a tuple & 2integer as input.the function should return a new tuple containing element from the original tuple within the specified range of indices.
def slice_tuple(tpl, start_index, end_index):
    """
    Return a new tuple containing elements from the original tuple within the specified range of indices.

    :param tpl: The original tuple.
    :param start_index: The start index (inclusive).
    :param end_index: The end index (exclusive).
    :return: A new tuple containing elements from the specified range.
    """
    # Ensure indices are within the bounds of the tuple
    if start_index < 0:
        start_index = 0
    if end_index > len(tpl):
        end_index = len(tpl)

    # Return the slice of the tuple
    return tpl[start_index:end_index]

# Example usage:
original_tuple = (10, 20, 30, 40, 50, 60, 70)
start = 2
end = 5
sliced_tuple = slice_tuple(original_tuple, start, end)
print(sliced_tuple)

#Q26.write a code that prompt a user to input 2 sets of character.then, print the union of these two sets.
def get_set_from_input(prompt):
    """
    Prompt the user for a set of characters separated by commas and return the set.

    :param prompt: The prompt message for user input.
    :return: A set of characters.
    """
    user_input = input(prompt)
    # Split the input string by commas and strip whitespace from each element
    characters = map(str.strip, user_input.split(','))
    # Return the set of characters
    return set(characters)

def main():
    """
    Main function to handle user input and print the union of two sets of characters.
    """
    print("Enter two sets of characters separated by commas.")

    # Get the first set from the user
    set1 = get_set_from_input("Enter the first set of characters: ")

    # Get the second set from the user
    set2 = get_set_from_input("Enter the second set of characters: ")

    # Find the union of the two sets
    union = set1 | set2

    # Print the result
    print("The union of the two sets is:", union)

# Run the main function
if __name__ == "__main__":
    main()


#Q27.develop a code that takes a tuple of  integer as input.the function should return the maximum & minimum values from the tuple using tuple unpacking.
def find_max_min(tpl):
    """
    Return the maximum and minimum values from a tuple of integers using tuple unpacking.

    :param tpl: A tuple of integers.
    :return: A tuple containing the maximum and minimum values.
    """
    # Check if the tuple is empty
    if not tpl:
        raise ValueError("The input tuple is empty.")

    # Find the maximum and minimum values using built-in functions
    max_value = max(tpl)
    min_value = min(tpl)

    # Return the results as a tuple
    return max_value, min_value

# Example usage:
input_tuple = (3, 5, 1, 9, 2)
max_value, min_value = find_max_min(input_tuple)
print(f"Maximum value: {max_value}")
print(f"Minimum value: {min_value}")


#Q28.create a code that define two sets of integer. then, print then union,intersection & difference of these two sets.
def main():
    """
    Define two sets of integers and print their union, intersection, and difference.
    """
    # Define two sets of integers
    set1 = {1, 2, 3, 4, 5}
    set2 = {4, 5, 6, 7, 8}

    # Compute union, intersection, and difference
    union = set1 | set2
    intersection = set1 & set2
    difference1 = set1 - set2  # Elements in set1 but not in set2
    difference2 = set2 - set1  # Elements in set2 but not in set1

    # Print the results
    print("Set 1:", set1)
    print("Set 2:", set2)
    print("Union:", union)
    print("Intersection:", intersection)
    print("Difference (set1 - set2):", difference1)
    print("Difference (set2 - set1):", difference2)

# Run the main function
if __name__ == "__main__":
    main()


#Q29.write a code that takes tuple & an element as input. the function should return the count of occurrence of the given element in the tuple.
def count_occurrences(tpl, element):
    """
    Return the count of occurrences of a given element in the tuple.

    :param tpl: The tuple in which to count occurrences.
    :param element: The element to count in the tuple.
    :return: The count of occurrences of the element.
    """
    # Use the count method of tuple to get the number of occurrences
    count = tpl.count(element)

    return count

# Example usage:
input_tuple = (1, 2, 3, 2, 4, 2, 5)
element_to_count = 2
occurrences = count_occurrences(input_tuple, element_to_count)
print(f"Element {element_to_count} occurs {occurrences} times in the tuple.")
def count_element_in_tuple(tpl, element):
    """
    Return the count of occurrences of a given element in the tuple.

    :param tpl: The tuple to search within.
    :param element: The element to count.
    :return: The count of occurrences of the element in the tuple.
    """
    # Use the count method to find the number of occurrences of the element
    count = tpl.count(element)

    return count

# Example usage:
input_tuple = (1, 2, 3, 4, 2, 2, 5, 6)
element_to_count = 2
count = count_element_in_tuple(input_tuple, element_to_count)
print(f"The element {element_to_count} appears {count} times in the tuple.")


#Q30.Develop a code that prompt the user to input 2 set of strings.then,print the symmetric difference of these two sets.
def get_set_from_input(prompt):
    """
    Prompt the user for a set of strings separated by commas and return the set.

    :param prompt: The prompt message for user input.
    :return: A set of strings.
    """
    user_input = input(prompt)
    # Split the input string by commas and strip whitespace from each element
    elements = map(str.strip, user_input.split(','))
    # Return the set of elements
    return set(elements)

def main():
    """
    Main function to handle user input and print the symmetric difference of two sets of strings.
    """
    print("Enter two sets of strings separated by commas.")

    # Get the first set from the user
    set1 = get_set_from_input("Enter the first set of strings: ")

    # Get the second set from the user
    set2 = get_set_from_input("Enter the second set of strings: ")

    # Compute the symmetric difference of the two sets
    symmetric_difference = set1 ^ set2

    # Print the result
    print("The symmetric difference of the two sets is:", symmetric_difference)

# Run the main function
if __name__ == "__main__":
    main()


#Q31.write a code that takes a list of words as input & return a dictionary where the keys are unique words & the values frequenceis of those words in input list.
def word_frequencies(words):
    """
    Return a dictionary with the frequency of each unique word in the input list.

    :param words: List of words.
    :return: Dictionary with words as keys and their frequencies as values.
    """
    frequency_dict = {}

    for word in words:
        # Increment the count for each word in the dictionary
        if word in frequency_dict:
            frequency_dict[word] += 1
        else:
            frequency_dict[word] = 1

    return frequency_dict

# Example usage:
input_list = ["apple", "banana", "apple", "orange", "banana", "apple"]
frequencies = word_frequencies(input_list)
print(frequencies)  # Output: {'apple': 3, 'banana': 2, 'orange': 1}

#Q32.write a code that takes two dictionaries as input & merge them into a single dictionary.if there are common keys, the value should be added together.
def merge_dictionaries(dict1, dict2):
    """
    Merge two dictionaries into a single dictionary. If there are common keys,
    the values are added together.

    :param dict1: The first dictionary.
    :param dict2: The second dictionary.
    :return: A new dictionary with merged keys and summed values.
    """
    # Create a new dictionary to hold the merged result
    merged_dict = dict1.copy()  # Start with a copy of the first dictionary

    # Iterate over the second dictionary
    for key, value in dict2.items():
        if key in merged_dict:
            # If the key exists in the merged dictionary, add the value
            merged_dict[key] += value
        else:
            # If the key does not exist, add it to the merged dictionary
            merged_dict[key] = value

    return merged_dict

# Example usage:
dict1 = {'a': 10, 'b': 20, 'c': 30}
dict2 = {'b': 15, 'c': 5, 'd': 10}
merged = merge_dictionaries(dict1, dict2)
print(merged)

#Q33.write a code to access a value in a nested dictionary.the function should take the dictionary & a list of keys as input, & return the corresponding value. If any of the keys do not exist in dictionary, the function should return none.
def get_nested_value(d, keys):
    """
    Access a value in a nested dictionary using a list of keys.
    Return None if any of the keys do not exist.

    :param d: The nested dictionary.
    :param keys: List of keys to navigate through the dictionary.
    :return: The value corresponding to the list of keys, or None if any key is missing.
    """
    # Start with the initial dictionary
    current_value = d

    # Iterate through each key in the list
    for key in keys:
        # Check if the key exists in the current dictionary
        if isinstance(current_value, dict) and key in current_value:
            current_value = current_value[key]
        else:
            # Return None if the key does not exist
            return None

    return current_value

# Example usage:
nested_dict = {
    'a': {
        'b': {
            'c': 10
        }
    },
    'x': 20
}

# List of keys to access
keys_to_access = ['a', 'b', 'c']
value = get_nested_value(nested_dict, keys_to_access)
print(value)

# Another example with a missing key
keys_to_access_missing = ['a', 'b', 'd']
value_missing = get_nested_value(nested_dict, keys_to_access_missing)
print(value_missing)

#Q34.write a code that takes a dictionary as input & return a sorted version of it based on the values.you can choose whether to sort in ascending or descending order.
def sort_dict_by_value(d, descending=False):
    """
    Return a sorted version of the dictionary based on its values.

    :param d: The dictionary to be sorted.
    :param descending: Boolean flag to sort in descending order (default is False for ascending).
    :return: A dictionary sorted by values.
    """
    # Sort the dictionary by values
    sorted_items = sorted(d.items(), key=lambda item: item[1], reverse=descending)

    # Convert the sorted items back to a dictionary
    sorted_dict = dict(sorted_items)

    return sorted_dict

# Example usage:
input_dict = {'a': 10, 'b': 5, 'c': 20, 'd': 15}

# Sort in ascending order
sorted_dict_asc = sort_dict_by_value(input_dict)
print("Sorted in ascending order:", sorted_dict_asc)

# Sort in descending order
sorted_dict_desc = sort_dict_by_value(input_dict, descending=True)
print("Sorted in descending order:", sorted_dict_desc)


#Q35.write a code that invert a dictionary, swapping keys & values.ensure that the inverted dictionary correctly handle cases where multiple keys have the same value by sorting the keys as list in the inverted dictionary.
from collections import defaultdict

def invert_dict(d):
    inverted_dict = defaultdict(list)

    for key, value in d.items():
        inverted_dict[value].append(key)

    # Convert defaultdict to a regular dict with sorted keys
    return dict(sorted(inverted_dict.items()))

# Example usage
original_dict = {
    'a': 1,
    'b': 2,
    'c': 1,
    'd': 3
}

inverted = invert_dict(original_dict)
print(inverted)
```
